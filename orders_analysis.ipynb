```python
# importing libraries
import pandas as pd
```

```python
# read data & handle null values
data = pd.read_csv("orders.csv", na_values=['Not Available', 'unknown'])
data.head(5)
```

```python
data['Ship Mode'].unique()
```

```python
# rename col names, lowercase and replacing space with underscores
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(" ","_")
data.columns
```

```python
data.head(5)
```

```python
# derive new columns = discount, sale price, profit
data["discount"]=data['list_price']*data['discount_percent']*0.01
data["sale_price"]=data['list_price'] - data['discount']
data["profit"]=data['sale_price'] - data['cost_price']
```

```python
data.head()
```

```python
# convert order_date from object datatype to datetime
data['order_date']=pd.to_datetime(data['order_date'], format = "%Y-%m-%d" )
```

```python
# drop irrelevant columns
data.drop(columns=['list_price','cost_price','discount_percent'], inplace = True)
```

```python
# load data to SQL Server
import sqlalchemy as sal
engine = sal.create_engine('mssql://Anubhav-PC\SQLEXPRESS/python_sql?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn = engine.connect()
```

```python
data.to_sql('data_orders', con=conn, index=False, if_exists='append')
```
